```python
df['Education'].value_counts()
```

```python
edu_dum = pd.get_dummies(df['Education'], drop_first=True, prefix='Education')
edu_dum.head()
```

```python
df = pd.concat([df, edu_dum], axis=1)
df.head()
```

```python
df = df.drop('Education', 1)
df.head()
```

```python
df['Profession'].value_counts()
```

```python
prof_dum = pd.get_dummies(df['Profession'], drop_first=True, prefix='Profession')
prof_dum.head()
```

```python
df = pd.concat([df, prof_dum], axis=1)
df = df.drop('Profession', 1)
df.head()
```

```python
df['Type of residence'].value_counts()
```

```python
res_dum = pd.get_dummies(df['Type of residence'], drop_first=True, prefix='Residence')
res_dum.head()
```

```python
df = pd.concat([df, res_dum], axis=1)
df = df.drop('Type of residence', axis=1)
df.head()
```
